```python
import numpy as np
import matplotlib.pyplot as plt

# Define the log-likelihood function for a normal distribution
def log_likelihood(mu, sigma_sq, x):
    n = len(x)
    return -n/2 * np.log(2 * np.pi * sigma_sq) - (1/(2 * sigma_sq)) * np.sum((x - mu)**2)

# Generate a range of mean values
mu_values = np.linspace(-10, 10, 100)

# Assume some observed data
x = np.random.normal(0, 1, 100)

# Calculate the log-likelihood for each mean value, for two different variances
log_likelihoods1 = [log_likelihood(mu, 0.1, x) for mu in mu_values]  # variance = 1
log_likelihoods2 = [log_likelihood(mu, 2, x) for mu in mu_values]  # variance = 2

# Plot the two curves
plt.figure(dpi=600)
plt.plot(mu_values, log_likelihoods1, label=r'$\sigma_0^2=0.1$')
plt.plot(mu_values, log_likelihoods2, label=r'$\sigma_0^2=2$')
plt.legend()
plt.xlabel(r'$\mu$')
plt.ylabel('log-likelihood')
plt.title(r'Log-likelihood of $\mu$ for different $\sigma_0^2$')
plt.show()
```

```python
# plot score function
plt.figure(dpi=600)
plt.plot(mu_values, np.gradient(log_likelihoods1), label=r'$\sigma_0^2=0.1$')
plt.plot(mu_values, np.gradient(log_likelihoods2), label=r'$\sigma_0^2=2$')
plt.legend()
plt.xlabel(r'$\mu$')
plt.ylabel('score function')
plt.title(r'Score function of $\mu$ for different $\sigma_0^2$')
plt.show()
```

```python
# generate 2 different samples
x1 = np.random.normal(0, 1, 10)
x2 = np.random.normal(0, 1, 10)
# plot their score functions
plt.figure(dpi=600)
plt.plot(mu_values, np.gradient([log_likelihood(mu, 0.1, x1) for mu in mu_values]), label=r'sample 1, $\sigma_0^2=0.1$')
plt.plot(mu_values, np.gradient([log_likelihood(mu, 0.1, x2) for mu in mu_values]), label=r'sample 2, $\sigma_0^2=0.1$')
plt.plot(mu_values, np.gradient([log_likelihood(mu, 2, x1) for mu in mu_values]), label=r'sample 1, $\sigma_0^2=2$')
plt.plot(mu_values, np.gradient([log_likelihood(mu, 2, x2) for mu in mu_values]), label=r'sample 2, $\sigma_0^2=2$')
plt.legend()
plt.xlabel(r'$\mu$')
plt.ylabel('score function')
plt.title(r'Score function of $\mu$ for different samples')
plt.show()
```
